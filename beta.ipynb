```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import StockHistory
import seaborn as sns
```

```python
names = StockHistory.get_namad_list()
```

```python
sec_data = pd.DataFrame()
asset =0
for n in names["NAMAD"]:
    d = StockHistory.get_namad_history_by_name(n,start_date='1393-1-1',end_date='1398-1-1',from_cache=True)["ClosePrice"]
    if len(d) >= 1200:
        asset+=1
        sec_data[n] = d
sec_data.dropna(inplace=True)        
sec_data.head()
```

```python
market_data = StockHistory.get_market_index(start_date='1393-1-1',end_date='1398-1-1')
market_data.drop(columns=["Date"],inplace=True)
```

```python
market_data.head()
```

```python
sec_returns = np.log(sec_data / sec_data.shift(1))
```

```python
sec_returns.head()
```

```python
market_data=np.log(market_data / market_data.shift(1))
market_data.head()
```

```python
sec_returns = sec_returns.merge(market_data,how='left',left_index=True,right_index=True)
sec_returns.head()
```

```python
res = []

for sec in sec_returns:
    cov_with_market = (sec_returns[[sec_returns[sec].name,"Value"]].cov()*20).iloc[0,1]
    market_variance = sec_returns["Value"].var() * 20
    beta = cov_with_market / market_variance
    res.append({"Name" : sec_returns[sec].name , "Beta" : beta})

beta_companies = pd.DataFrame(res).set_index("Name")
```

```python
beta_companies[(beta_companies["Beta"] > 0) & (beta_companies["Beta"] < 1)].sort_values("Beta",ascending = True)
```
